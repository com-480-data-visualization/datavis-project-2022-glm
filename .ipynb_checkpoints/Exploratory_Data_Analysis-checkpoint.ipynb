# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
wec = pd.read_csv('World Energy Consumption.csv')
lon_lat = pd.read_csv('world_country_latitude_and_longitude_values.csv')[['latitude', 'longitude', 'country']]

# Two changes are necessary in order to have a perfect matching between the names of the countries from both datasets
lon_lat = lon_lat.replace('Congo [DRC]', 'Democratic Republic of Congo')
lon_lat = lon_lat.replace('Congo [Republic]', 'Congo')
```

Here we can see all the columns that are in the world energy consumption dataset, so we can see what information is covered.

```python
for col in wec.columns:
    print(col)
```

We can see that there are a lot of statistics about each source of energy. However, some of these measures are redundant with others. For example, the information on low carbon energy is the addition of the information from renewables and nuclear, which we will probably consider separately. 

Then, there is some data that we might not need if we keep some sufficient information such as:

* The consumption per country for each source which corresponds to the primary energy that is consumed. For example, a liter of oil that is burning can deliver 9.87 kWh.
* The electricity per coutry from each source, which correponds to the energy that is actually supplied to the grid. As the electricity is generated by the consumption from primary energy, it is always lower than consumption as shown in the next section.
* Other information such as the name of the coutry and related information such as the population or the grand domestic product that might be useful.

Below, we can see the list of the countries on which we have data in the world energy consumption dataset.

```python
for country in wec.country.unique():
    print(country)
```

Below, we can see the mean of energy consumption from each source, which can give us some insight about their shares in energy consumption.

```python
gas = wec['gas_consumption'].mean()
oil = wec['oil_consumption'].mean()
coal = wec['coal_consumption'].mean()
nuclear = wec['nuclear_consumption'].mean()
renew = wec['renewables_consumption'].mean()

print(gas)
print(oil)
print(coal)
print(nuclear)
print(renew)
```

## Difference between consumption and electricity

Here we illustrate the difference between consumption and electricity features in the data.

```python
france_ec = wec[wec['country'] == 'France']

year = france_ec['year']
prod = france_ec['nuclear_electricity']
cons = france_ec['nuclear_consumption']

plt.title('Energy (in TWh) from nuclear power plants in France')
plt.plot(year, prod, 'g', label= 'Electricity')
plt.plot(year, cons, 'r', label= 'Consumption')
plt.legend()
plt.show()
```

We can see the curb of the electricity is correlated to the consumption, but it is smaller probably due to the loss during the process that is operated. Most of the energy from nuclear power plants is used to supply the grid and therefore to generate electricity, but this is not the case for all sources as we can see below.

```python
prod = france_ec['oil_electricity']
cons = france_ec['oil_consumption']

plt.title('Energy (in TWh) from oil in France')
plt.plot(year, prod, 'g', label= 'Electricity')
plt.plot(year, cons, 'r', label= 'Consumption')
plt.legend()
plt.show()
```

We can see that there is almost no electricity generated from oil. That is because most of the oil is used by transports such as cars, and rarely used to supply electricity to the grid. 

The graph below shows the difference between elctricity and consumption for renewables, and we can see that they are correlated like the case with the nuclear plants. In this case, we can suppose that renewables are mostly used to supply the grid and that the difference is mostly due to loss during the generation process.

```python
prod = france_ec['renewables_electricity']
cons = france_ec['renewables_consumption']

plt.title('Energy (in TWh) from renewables in France')
plt.plot(year, prod, 'g', label= 'Electricity')
plt.plot(year, cons, 'r', label= 'Consumption')
plt.legend()
plt.show()
```

## Data on renewables

Now, we want to give more focus on the renewables and this is why we are looking to the countries on which we have information on renewables electricity and consumption.

```python
#join the 2 dataframes on country
wec = wec.merge(lon_lat)
```

```python
columns_electricity = ['hydro_electricity', 'solar_electricity', 'wind_electricity',
                       'renewables_electricity', 'biofuel_electricity']                    
columns_renewables = ['renewables_consumption', 'hydro_consumption', 'solar_consumption', 'wind_consumption', 
                      'biofuel_consumption']
column_info = ['iso_code', 'country', 'year', 'latitude', 'longitude', 'gdp', 'population']
wec_all = wec[column_info + columns_electricity + columns_renewables]
```

```python
wec_all
```

```python
#number of countries
len(wec_all.country.unique())
```

```python
wec_all.isna().sum()
```

Naturally, data may not available for all of the 204 countries present in the dataset, so let's see what are the countries for which data is available. We will split the analysis in two: data related to electricity and data related to consumption.

```python
wec_consumption = wec[column_info + columns_renewables]
#drop rows with NaNs
wec_consumption = wec_consumption.dropna(subset=columns_renewables)
#nr of countries
len(wec_consumption.country.unique())
```

Let's see which are these countries.

```python
#visualize the countries
fig = px.scatter_mapbox(wec_consumption, lat="latitude", lon="longitude", hover_name="country", zoom=0)
fig.update_layout(mapbox_style="open-street-map")
fig.show()
```

```python
wec_electricity = wec[column_info + columns_electricity]
#drop rows with NaNs
wec_electricity = wec_electricity.dropna(subset=columns_electricity)
#nr of countries
len(wec_electricity.country.unique())
```

Electricity related data is available for most of the countries, but probably the time span is small (data is only available for a few years). Let's verify this.

```python
country_years_count = wec_electricity.groupby('country').year.count().reset_index(name='years_count')
country_years_count
```

```python
#years available and number of countries
country_years_count.years_count.value_counts()
```

The hypothesis is therefore true, most of the countries (135/202) have 20 years of electricity data available, which should suffice for our analysis.
